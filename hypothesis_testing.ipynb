### Perform hypothesis testing using COVID-19 Indonesia dataset

Dataset source: https://www.kaggle.com/datasets/hendratno/covid19-indonesia

```python
# Import library used
import pandas as pd
from scipy import stats
```

```python
# Import dataset used
covidINA = pd.read_csv("covid_19_indonesia_time_series.csv")
```

```python
# Expand the output display to see more columns
pd.set_option('display.max_columns', None)
```

```python
covidINA
```

```python
# Make new dataframe covidDate using Date for the index
covidDate = covidINA.set_index('Date').sort_values('Date')
```

```python
covidDate
```

```python
# Showing unique value of Location to chose province for testing
covidDate['Province'].unique()
```

Hypothesis Testing using ANOVA

we will test whether the mean of total cases of DKI Jakarta, Jawa Timur, and Sumatera Selatan are significantly different or not.

```python
covidJakarta = covidDate[covidDate.Province == 'DKI Jakarta'].groupby('Date').sum(numeric_only=False)['Total Cases']
covidJawaTimur = covidDate[covidDate.Province == 'Jawa Timur'].groupby('Date').sum(numeric_only=False)['Total Cases']
covidSumateraSelatan = covidDate[covidDate.Province == 'Sumatera Selatan'].groupby('Date').sum(numeric_only=False)['Total Cases']

print("Daily Average of DKI Jakarta covid case",covidJakarta.mean().round())
print("Daily Average of Jawa Timur covid case",covidJawaTimur.mean().round())
print("Daily Average of Sumatera Selatan covid case",covidSumateraSelatan.mean().round())
```

It seems that they are significantly different since `DKI Jakarta` average is `587520` case per day and the rest is between half of DKI Jakarta `38019` for `Sumatera Selatan` and `263062` for `Jawa Timur`.

```python
f_stat,p_value = stats.f_oneway(covidJakarta, covidJawaTimur, covidSumateraSelatan)
print('P-value:',p_value)
```

**Since the p-value is below 0.05, then we reject the Null Hypothesis.** We conclude that the difference of DKI Jakarta, Jawa Timur, and Sumatera Selatan is statistically significant.
